4. Describe how dictionaries store data.
Dictionaries in Python store data as key-value pairs. Each key is unique, and it maps to a specific value. Dictionaries are defined using curly braces {}.

python
my_dict = {'name': 'Alice', 'age': 25}
# Access value by key: my_dict['name']  # Output: 'Alice'

7. How do tuples ensure data integrity in Python?
Tuples ensure data integrity by being immutable. Once a tuple is created, its content cannot be changed. This characteristic makes tuples suitable for storing data that should not be modified.

8. What is a hash table, and how does it relate to dictionaries in Python?
A hash table is a data structure that maps keys to values using a hash function. In Python, dictionaries are implemented using hash tables, allowing for fast data retrieval based on keys.

9. Can lists contain different data types in Python?
Yes, lists can contain elements of different data types in Python.

10. Explain why strings are immutable in Python.
Strings are immutable in Python to ensure that they remain consistent and to allow for optimizations like interning, where identical strings are stored only once in memory.

11. What advantages do dictionaries offer over lists for certain tasks?
Dictionaries offer faster lookups for data retrieval using keys, making them ideal for tasks requiring quick access to data based on unique identifiers.

12. How do sets handle duplicate values in Python?
Sets automatically remove duplicate values when they are created or when elements are added.

13. Describe a scenario where using a tuple would be preferable over a list.
Using a tuple would be preferable when you want to ensure that the data remains unchanged, such as storing coordinates, fixed configurations, or as keys in a dictionary.

14. How does the “in” keyword work differently for lists and dictionaries?
For lists, the in keyword checks if an element is present in the list.

python
my_list = [1, 2, 3]
2 in my_list  # Output: True
For dictionaries, the in keyword checks if a key is present in the dictionary.

python
my_dict = {'a': 1, 'b': 2}
'a' in my_dict  # Output: True
15. Can you modify the elements of a tuple? Explain why or why not.
No, you cannot modify the elements of a tuple because tuples are immutable. This immutability ensures data integrity and consistency.

16. What is a nested dictionary, and give an example of its use case.
A nested dictionary is a dictionary that contains another dictionary as a value.

python
nested_dict = {
    'user1': {'name': 'Alice', 'age': 25},
    'user2': {'name': 'Bob', 'age': 30}
}
# Accessing nested value: nested_dict['user1']['name']  # Output: 'Alice'
Use Case: Storing complex data structures like user profiles in a database.

17. Describe the time complexity of accessing elements in a dictionary.
The average time complexity of accessing elements in a dictionary is O(1) (constant time), thanks to the hash table implementation.

18. In what situations are lists preferred over dictionaries?
Lists are preferred when you need ordered collections of elements or when you need to store multiple items that may contain duplicates.

19. Why are dictionaries considered unordered, and how does that affect data retrieval?
Dictionaries are considered unordered because they do not maintain the order of insertion. Data retrieval is based on keys rather than order, making it efficient for lookup operations.
